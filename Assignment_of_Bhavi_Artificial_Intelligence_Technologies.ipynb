**Metrics Calculation and Export**

```
class SMAStrategy:
    # ... (entry and exit strategy methods remain the same)

    def export_trades_to_csv(self):
        trades_df = pd.DataFrame(self.trades)
        trades_df.to_csv('trades.csv', index=False)

    def calculate_metrics(self):
        total_profit_loss = sum(trade['profit_loss'] for trade in self.trades)
        success_rate = sum(trade['profit_loss'] > 0 for trade in self.trades) / len(self.trades)
        total_trades = len(self.trades)
        # Print or store the calculated metrics
        print(f"Total Profit/Loss: {total_profit_loss}")
        print(f"Success Rate: {success_rate}")
        print(f"Total Trades: {total_trades}")
```
